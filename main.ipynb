#### Catogrizing the Reviews 

```python
# Define a function to catgorize the rating 
def categorize_rating(rating):
    if rating >= 4.0:
        return 'Positive'
    elif rating >= 3.0:
        return 'Neutral'
    else:
        return 'Negative'

# applying function to create a new 'Category' column
ola_df['rating_type'] = ola_df['rating'].apply(categorize_rating)
uber_df['rating_type'] = uber_df['rating'].apply(categorize_rating)
```

#### Normlizing the text

```python
# Convert the review_description to a string
ola_df['review_description'] = ola_df['review_description'].astype(str)
uber_df['review_description'] = uber_df['review_description'].astype(str)
```

```python

import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Function to preprocess a text string
def preprocess_text(text):
    # Convert to lowercase
    lower_string = text.lower()

    # Remove numbers
    no_number_string = re.sub(r'\d+', '', lower_string)

    # Remove all punctuation except words and space
    no_punc_string = re.sub(r'[^\w\s]', '', no_number_string)

    # Remove white spaces
    no_wspace_string = no_punc_string.strip()

    # Convert string to list of words
    lst_string = [no_wspace_string][0].split()

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    no_stpwords_string = " ".join(word for word in lst_string if word not in stop_words)

    return no_stpwords_string


```

```python
# Apply the preprocessing function to the 'review_description' column
ola_df['normalized_review_description'] = ola_df['review_description'].apply(preprocess_text)
uber_df['normalized_review_description'] = uber_df['review_description'].apply(preprocess_text)
# Display the DataFrame with the normalized review descriptions
ola_df[['review_description', 'normalized_review_description']].head(2)
```

# 5-Modeling 

### creating topics from the negative reviews from ola reviews  

```python
from gensim import corpora
from gensim.models import LdaModel
from gensim.parsing.preprocessing import preprocess_string



# Preprocess the text data

documents = n_user_review['normalized_review_description'].dropna().tolist()

# Tokenize and preprocess the text
processed_docs = [preprocess_string(doc) for doc in documents]

# Create a dictionary and a corpus
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

# Build the LDA model
lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15, random_state=42)

# Print the topics and associated words
topics = lda_model.print_topics(num_words=10)
for topic in topics:
    print(topic)
```

```python
topics = lda_model.print_topics(num_words=10)

# Create a bar plot for each topic
for topic_id, topic in topics:
    words = [word.split("*")[1].strip(' "') for word in topic.split(" + ")]
    word_probs = [float(word.split("*")[0]) for word in topic.split(" + ")]
    
    plt.figure(figsize=(8, 6))
    plt.barh(words, word_probs)
    plt.xlabel('Word Probability')
    plt.title(f'Top Words in Topic {topic_id}')
    plt.gca().invert_yaxis()  
    plt.show()
```

```python
negative_reviews_ola.columns
```

```python
# Create a new column to store assigned topics
negative_reviews_ola['Assigned Topics'] = ''

for index, row in negative_reviews_ola.iterrows():
    review = row['normalized_review_description']
    doc_bow = dictionary.doc2bow(review.split())
    topic_distribution = lda_model[doc_bow]
    
    # Sort topics by probability and select the top one
    top_topic = sorted(topic_distribution, key=lambda x: x[1], reverse=True)[0]
    
    # Assign the topic number to the 'Assigned Topics' column
    negative_reviews_ola.at[index, 'Assigned Topics'] = top_topic[0]
```
